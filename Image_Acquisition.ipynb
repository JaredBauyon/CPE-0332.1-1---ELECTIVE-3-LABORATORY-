```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/orange.png')
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_size = cv2.resize(img_gray, (300, 300))
cv2_imshow(img_size)
```

```python
img_flip = cv2.flip(img_size, -1
                    )
cv2_imshow(img_flip)
```
